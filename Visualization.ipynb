```python
# Gather Data
csvFile = "sample_csv.csv"
data = pd.read_csv(csvFile);
data
```

```python
# station object
class Station():
    def __init__(self, station_id, name="Unkown", lat=0, long=0, lines=[], neighbors=[]):
        self.station_id = station_id
        self.name = name
        self.lat = lat
        self.long = long
        self.lines = lines
        self.neighbors = []
    def __lt__(self, other):
        return 1
    def __eq__(self, other):
        if self.station_id == other.station_id:
            return True
    def __repr__(self):
        return self.station_id
        
class trip():
    def __init__(self, trip_id, start, end, transit_time, walking_time, public_time):
        pass
```

```python
# instatiate stations and lines
stations = []
lines = set()

with open(csvFile, newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        # skip title row
        if(row[0] == "ID"):
            continue
            
        # init lines used
        lines_used = set(row[4])
        for line in lines_used:
            lines.add(line)

        # init station
        stations.append(Station(int(row[0]), row[1], float(row[2]), float(row[3]), list(row[4])))
```

```python
# visualize grouping by line
from pyvis import network as net


net = net.Network(directed=True, notebook=True)

# create nodes for stations
for station in stations:
    net.add_node(station.station_id, station.name)
    
# create nodes for lines
net.add_nodes(list(lines))
    
# add edges to stations
for station in stations:
    pass
    for line in station.lines:
        net.add_edge(station.station_id, line)


net.show('nx.html')
```
